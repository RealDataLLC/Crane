```python
import pandas as pd
from pathlib import Path

top_level_csv_files = Path("drawdown-master/solution/")
aloc = Path("drawdown-master/data/")

```

## Getting all the vma data

```python
data = pd.DataFrame()

for i in top_level_csv_files.glob('*/vma_data/*.csv'):
    #print(i.name)
    if i.name != "VMA_info.csv":
        tmp = pd.read_csv(i)
        tmp["Category"] = i.parent.parent.name
        tmp["Directory"] = i
        data = pd.concat([data,tmp], axis = 0)
```

```python
data.head()
```

```python
data.to_csv("Solutions_VMA_compiled.csv")
```
