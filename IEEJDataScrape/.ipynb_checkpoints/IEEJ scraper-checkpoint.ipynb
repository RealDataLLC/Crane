```python
def fillBlanks(table, year_row = 0, columns = 0):
    for i in range(year_row, len(table)): 
        # get errors in column names up to 999,999,999
        try:
            numbers = re.findall("-?\d*[,.]?\d*[,.]?\d+%?-?", str(table.iloc[i,columns]))

            if len(numbers) > 1:
                try:
                    text = re.sub(numbers[0], "", table.iloc[i,columns])
                    table.iloc[i,columns] = text
                    table.iloc[i,columns+1] = numbers[0]
                except:
                    print(numbers, table.iloc[i,columns])
        except:
            print(i)
            
    return table
```

```python
def fillNegBlanks(table, year_row = 0, columns = 0):
    for i in range(year_row, len(table)): 
        # get errors in column names up to 999,999,999
        try:
            numbers = re.findall("(-) (\d*[,.]?\d*[,.]?\d+%?-?)", str(table.iloc[i,columns]))[0]
            if len(numbers) > 1:
                try:
                    table.iloc[i,columns] = numbers[0]
                    table.iloc[i,columns+1] = numbers[1]
                except:
                    print(numbers, table.iloc[i,columns])
        except:
            print(i)
            
    return table
```

```python
ex_page = '30' # 2 to 38

# area
left = 48
right = 48 + 261

#columns
columns = [131, 157, 181, 207, 233, 262, 286, 308]
```

```python
# by page
ref_geo = ["World", "Asia", "China", "India", "Japan", "Korea", "Chinese Taipei", "ASEAN", "Indonesia", "Malaysia", "Myanmar", "Phillipines", "Thailand", "Vietnam", "North America", "United States", "Latin America", "OECD Europe", "Non-OECD Europe", "European Union", "Africa", "Middle East", "Oceania", "OECD", "Non-OECD"]
# by page
adv_geo = ["World", "Asia", "China", "India", "Japan", "ASEAN", "United States", "European Union"]
# by page
sub_geo = ["World", "Asia"]
# by page
scenarios = ["Reference", "Advanced Technologies", "No New Coal Plants (Natural Gas Substitution)", "No New Coal Plants (Renewables Substitution)"]
# by table
groups = ["Primary Energy Consumption", "Final Energy Consumption", "Electricity Generation"]
# by table
unit = ["Mtoe", "TWh"]
```

```python
table1 = tabula.read_pdf('IEEJfixed.pdf', pages = ex_page,multiple_tables=True)#, columns= columns)
table1
```

```python
doc_table = pd.DataFrame()
for page in range(2, 35):
    
    table1 = tabula.read_pdf('IEEJfixed.pdf', pages = page, multiple_tables=True)#, columns= columns)

    #Parameter addition for each of the tables
    total = pd.DataFrame()
    # -1 becase we don't care about the last table
    for i in range(len(table1)-1):
        sliced = table1[i].iloc[:,:8]
        if i == 2:
            sliced["group"] = groups[2]
            sliced["units"] = unit[1]
        elif i == 1:
            sliced["group"] = groups[1]
            sliced["units"] = unit[0]
        elif i == 0:
            sliced["group"] = groups[0]
            sliced["units"] = unit[0]
        else:
            print("something wrong happened")
        total = pd.concat([total, sliced], axis = 0)
    page_table = total.dropna()
    #print(page)
    assert len(page_table) == 34

    #Parameter addition based on each of the pages
    if page in range(2, 27):
        page_table["scenario"] = scenarios[0]
        page_table["geography"] = ref_geo[page-2]
    elif page in range(27, 35):
        page_table["scenario"] = scenarios[1]
        page_table["geography"] = adv_geo[page-27]
    elif page in range(35, 37):
        page_table["scenario"] = scenarios[2]
        page_table["geography"] = sub_geo[page-35]
    elif page in range(37, 39):
        page_table["scenario"] = scenarios[3]
        page_table["geography"] = sub_geo[page-37]
    else:
        print("something wrong")
    
    #adding totals
    doc_table = pd.concat([doc_table, page_table], axis = 0)
```
